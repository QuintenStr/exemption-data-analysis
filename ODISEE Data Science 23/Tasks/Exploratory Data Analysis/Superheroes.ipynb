**Maak de gecombineerde dataframe df_all_heroes van de stats, info en de powerMatrix. Gebruik hiervoor de merge functie. Op welke kolom ga je de join uitvoeren? Om welke reden heb je voor deze kolom gekozen?**

Antwoord: Op de **naam**. Dit is het meest logische en het aspect dat gemeenschappelijk is tussen de drie tabellen.

Let ervoor op dat je rekening houdt met de volgende zaken: 
* Zorg ervoor dat alle helden die in de powermatrix en stats dataframes aanwezig zijn (outer join)
* Voeg daarna de bijhorende kolommen toe uit het info dataframe (met een inner join)
* Let erop dat boolean kolommen nog steeds boolean zijn. Indien dit niet zo is, moet je het data type veranderen van bool naar boolean (zodat het NaN waarden kan bevatten en dus nullable is).

Als resultaat zou je een dataframe met een 900-tal entries en 190-tal kolommen moeten hebben.

```python
# merge dataframes

df_all_heroes = df_superheroes_power_matrix.merge(df_superheroes_stats, on='Name', how='outer', suffixes=('_left', '_right')).merge(df_superheroes_info, on='Name', how='inner', suffixes=('_left', '_right'))

list_bool_column_names = []

for (columnName, columnData) in df_superheroes_power_matrix.iteritems():
    if(columnData.dtypes == 'bool'):
        list_bool_column_names.append(f'{columnName}')
print(list_bool_column_names)

# gevonden dat sommige kolommen overlappen 
# df2 = df_all_heroes.loc[:, list_bool_column_names] 
# geeft: KeyError: "['Durability', 'Intelligence'] not in index"
# dit wil zeggen dat hij Durability en Intelligence niet vindt. Dit komt omdat er een _left is toegevoegd, omdat je een Durability en Intelligence hebt in zowel de power matrix als stats tabel. De kolommen die wij willen aanspreken zijn die van uit de power matrix,
# dus vervang ik even durability door durability_left en idem voor intelligence.

for i in range(len(list_bool_column_names)):
    if list_bool_column_names[i] == 'Durability':
        list_bool_column_names[i] = 'Durability_left'
  
    if list_bool_column_names[i] == 'Intelligence':
        list_bool_column_names[i] = 'Intelligence_left'

df_all_heroes[list_bool_column_names] = df_all_heroes[list_bool_column_names].astype('bool')

display(df_all_heroes)
display(df_all_heroes.dtypes)
# als je deze 2 lijnen uitvoert net na de merge zie je dat de bools objects zijn geworden en nu op het einde dus weer succesvol omgevormd is naar bools
```

Zet de "Name" kolom van de superhelden als index van het dataframe.
Normaal zijn er ook twee kolommen met dezelfde Alignment voorzien (standaard met de suffixes _x en _y).
Drop 1 van de twee kolommen en verwijder de suffix van de andere kolom zodat de uiteindelijke naam Alignment is.

```python
df_all_heroes = df_all_heroes.set_index('Name')

# kleine check om te zien of er nog andere kolommen zijn die dubbel voorkomen
col_list = [col for col in df_all_heroes.columns if col.endswith('_left') or col.endswith('_right')]
print(col_list)
# zoals vorige oefening; durability en intelligence zijn dubbel en dan inderdaad intelligence ook.

# 'Alignment_left' droppen
df_all_heroes = df_all_heroes.drop(['Alignment_left'], axis=1)
display(df_all_heroes)
```

Er zijn drie soorten kolommen in het dataframe aanwezig, namelijk numerieke, boolean en tekstuele kolommen.
We gaan nu de NaN-waarden invullen (imputen) in het dataframe door de volgende stappen uit te voeren met behulp van de fillna methode:
* NaN waarden in een kolom met boolean data kan je vervangen door False.
* NaN waarden in een kolom met numerieke data kan je vervangen door het gemiddelde
* NaN waarden in tekstuele kolommen kan je vervangen door het keyword "Unknown"

**Doe dit nu en controleer het aantal null values dat overblijft.**

```python
# fill NaN values
# boolean opvullen
df_all_heroes[df_all_heroes.select_dtypes(include='bool').columns] = df_all_heroes[df_all_heroes.select_dtypes(include='bool').columns].fillna('False')

# numeric opvullen
df_mean_numeric = df_all_heroes.select_dtypes('number').describe().loc['mean']
dict_mean_numeric = df_mean_numeric.to_dict()
df_all_heroes[df_all_heroes.select_dtypes(include='number').columns] = df_all_heroes[df_all_heroes.select_dtypes(include='number').columns].fillna(value=dict_mean_numeric)

# string (object) opvullen
df_all_heroes[df_all_heroes.select_dtypes(include=object).columns] = df_all_heroes[df_all_heroes.select_dtypes(include=object).columns].fillna('Unknown')

print(f'Aantal resterende null waarden: {df_all_heroes.isnull().sum().sum()}')
display(df_all_heroes)


# voor latere oefening: selectie met numerieke data maken
numerieke_data_kolomnamen = df_all_heroes.select_dtypes(include='number').columns
```

Nu gaan we nog twee zaken veranderen in de dataset om de privacy van de superhelden te verzekeren.
Een eerste stap is om variantie van +- 5% toe te voegen aan alle kolommen die floating point waarden bevatten.
Daarnaast gaan we ook de indices door elkaar shuffelen zodat de naam van een superheld niet meer gekoppeld is aan zijn krachten.

**Voer nu variantie toe en shuffle de namen voor de privacy te garanderen. Sla de resulterende dataset op als superheroes-info-and-stats/processedData.csv**

```python
# data anonymisation
selectie_float = df_all_heroes.select_dtypes(include=['float64'])

percentage = 5
random_waarde = random.uniform((1-(percentage/100)),(1+(percentage/100)))
print('Voor de 5% variatie')
display(selectie_float)

for i in range(selectie_float.shape[0]):
    for j in range(selectie_float.shape[1]):
            selectie_float.iat[i,j] *= random_waarde

print('Na de 5% variatie')
display(df_all_heroes.select_dtypes(include=['float64']))


# namen shuffelen
df_all_heroes_reset_index = df_all_heroes.reset_index()
df_all_heroes_reset_index['Name'] = np.random.permutation(df_all_heroes_reset_index['Name'])
df_all_heroes = df_all_heroes_reset_index.set_index('Name')

# duplicates eruit halen om ervoor te zorgen dat als je random 20 neemt voor plots te maken volgende oefening niet 2 keer dezelfde index hebt

df_all_heroes = (df_all_heroes.reset_index().drop_duplicates(subset='Name').set_index('Name'))
print('Na namen te shuffelen')
display(df_all_heroes)

df_all_heroes.to_csv(index=True, path_or_buf=".\superheroes-info-and-stats\processedData.csv")
# hou er rekening mee dat als je dit rerunt, je 5% varatie gaat toevoegen op de 5% variatie die er al in zat, dus als je het wil testen run alles opnieuw vanaf de merge oefening in respectievelijke volgorde!!
```
